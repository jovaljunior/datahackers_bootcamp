```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

plt.style.use('ggplot')
```

```python
df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
```

```python
plt.figure(figsize=(17, 10))
sns.heatmap(df_train.corr())
plt.show()
```

```python
df_train.corr()['SalePrice'].sort_values(ascending=False)
```

```python
df_train = df_train.select_dtypes(include=np.number)
df_test = df_test.select_dtypes(include=np.number)
```

```python
df_train.isnull().sum().sort_values(ascending=False)
```

```python
df_train = df_train.interpolate()
df_test = df_test.interpolate()
```

```python
df_train['SalePrice'] = np.log(df_train['SalePrice'])
```

```python
df_train.head()
```

```python
df_train.drop('Id', axis=1, inplace=True)
```

```python
X = df_train.drop('SalePrice', axis=1)
y = df_train['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.33)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train)
```

```python
print('Score no conjunto de treino: {:.2%}'.format(reg.score(X_train, y_train)))
print('Score no conjunto de teste: {:.2%}'.format(reg.score(X_test, y_test)))
```

```python
df_test['SalePrice'] = np.exp(reg.predict(df_test.drop('Id', axis=1)))
```

```python
submission = df_test[['Id', 'SalePrice']]
submission.head()
submission.to_csv('submission.csv', index=False)
```
