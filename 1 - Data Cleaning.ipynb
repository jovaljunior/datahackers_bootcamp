```python
import pandas as pd
```

```python
df = pd.read_csv("ceaps_dataset.csv", skiprows=1, delimiter=";", encoding="latin1")
df.head()
```

```python
df['VALOR_REEMBOLSADO'] = df['VALOR_REEMBOLSADO'].str.replace(',','.')
df['VALOR_REEMBOLSADO'] = pd.to_numeric(df['VALOR_REEMBOLSADO'])
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['DATA'] = pd.to_datetime(df['DATA'], errors='coerce')
```

```python
df.dtypes
```

```python
df[df['DATA'].isnull()]
```

```python
df.dropna(subset=['DATA'], inplace=True)
```

```python
df.head()
```

```python
df['DATA'].dt.day
```

```python
df['DIA'] = df['DATA'].dt.day
```

```python
df.head()
```

```python
df['TIPO_DESPESA'].unique()
```

```python
mapper = {'Aluguel de imóveis para escritório político, compreendendo despesas concernentes a eles.' : 'Aluguel', 'Aquisição de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos. ': 'Aquisição de Materia', 'Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar': 'Contratações', 'Locomoção, hospedagem, alimentação, combustíveis e lubrificantes': 'Logística', 'Passagens aéreas, aquáticas e terrestres nacionais': 'Passagens', 'Divulgação da atividade parlamentar': 'Divulgações', 'Serviços de Segurança Privada': 'Segurança'}
```

```python
df['TIPO_DESPESA'] = df['TIPO_DESPESA'].map(mapper)
```

```python
df.head()
```

```python
df['TIPO_DESPESA'] = df['TIPO_DESPESA'].str.replace('Aquisição de Materia', 'Aquisição de Material')
```
